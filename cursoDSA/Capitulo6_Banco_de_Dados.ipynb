```python
# Importando o módulo de acesso ao SQLite
import sqlite3
import datetime
import random
import matplotlib.pyplot as plt
%matplotlib notebook

# Cria uma conexão com o banco de dados.
conn = sqlite3.connect('dsa.db')

# Criando um cursor
c = conn.cursor()
```

```python
#Função para criar uma tabela:
def create_table():
    c.execute('CREATE TABLE IF NOT EXISTS produtos(id INTEGER PRIMARY KEY AUTOINCREMENT, date TEXT, '\
              'prod_name TEXT, valor REAL)')
```

```python
#Função para inserir em uma tabela:
def insert():
    c.execute("INSERT INTO produtos (date, prod_name, valor) VALUES('2018-01-02 10:02:11', 'Mouse', 100)")
    conn.commit()
```

```python
# Usando variáveis para inserir dados    
def insert_var():
    new_date = datetime.datetime.now()
    new_prod_name = 'Teclado'
    new_valor = random.randrange(50,100)
    c.execute("INSERT INTO produtos (date, prod_name, valor) VALUES (?, ?, ?)", (new_date, new_prod_name, new_valor))
    conn.commit()
```

```python
# Buscando dados
def select_dados():
    c.execute("SELECT * FROM PRODUTOS")
    for linha in c.fetchall():
        print(linha)
```

```python
# Atualiza os dados
def atualiza_dados():
    c.execute("UPDATE produtos SET valor = 70.00 WHERE valor > 80.0")
    conn.commit()
```

```python
# Deleta os dados
def remove_dados():
    c.execute("DELETE FROM produtos WHERE valor = 65.0")
    conn.commit()
```

```python
# Gerar gráficos com os dados        
def dados_grafico():
    c.execute("SELECT id, valor FROM produtos")
    ids = []
    valores = []
    dados = c.fetchall()
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])
        
    plt.bar(ids, valores)
    plt.show()
```
